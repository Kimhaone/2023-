```python
from statsmodels.formula.api import ols

Rvar = '특수학교 ~ 장애인인구수_비중'
regression_result = ols(Rvar, data = df).fit()
```

```python
regression_result.summary()

#R제곱, 즉 결정계수가 0.003으로 두 변수간의 낮은 상관관계를 보임
#유의수준이 높게 나와서 변수로서 적합하지 않음
```

```python
from statsmodels.formula.api import ols

Rvar = '국내대학진학수 ~ 특수학급_고등'
regression_result = ols(Rvar, data = df).fit()

regression_result.summary()
```

```python
df1 = df[['시군구명','국내대학진학수','특수학급_고등']]
df1.set_index('시군구명')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
rg = LinearRegression()
rg_x = df1[['특수학급_고등']]
rg_y = df1[['국내대학진학수']]
rg.fit(rg_x, rg_y)
```

```python
prediction = rg.predict(rg_x)
df['est.uiv_rates']=prediction
df.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False

plt.scatter(df[['특수학급_고등']], df[['국내대학진학수']])
plt.plot(rg_x, prediction, color = 'blue', linewidth=5)
plt.xlabel('자치구별 특수학급 수_고등')
plt.ylabel('자치구별 국내대학진학자 수')
plt.show()
```
